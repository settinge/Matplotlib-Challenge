```python
mouse_metadata = "C:\\Users\\samantha.ettinger\\GWU-ARL-DATA-PT-12-2019-U-C\\02-Homework\\05-Matplotlib\\Instructions\\Pymaceuticals\\data\\Mouse_metadata.csv"
study_results = "C:\\Users\\samantha.ettinger\\GWU-ARL-DATA-PT-12-2019-U-C\\02-Homework\\05-Matplotlib\\Instructions\\Pymaceuticals\\data\\Study_results.csv"
```

```python
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)
```

```python
merge_table = pd.merge(study_results, mouse_metadata, on="Mouse ID")
```

```python

# * Generate a summary statistics table consisting of the mean, median, variance, standard deviation, and SEM of the tumor volume for each drug regimen.
mean_table=merge_table.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
mean_table
median_table=merge_table.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
median_table
var_table=merge_table.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
var_table
sd_table=merge_table.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
sd_table
std_error_table=merge_table.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']
std_error_table

statistic_summary_df=pd.DataFrame({'Mean':mean_table,'Median':median_table,'Variance':var_table,'Standard Deviation':sd_table,'SEM':std_error_table})
statistic_summary_df
```
